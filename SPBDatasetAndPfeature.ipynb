```python
# Amino acid composition (AAC)

from Pfeature.pfeature import aac_wp

def aac(input):
  a = input.rstrip('fasta')  #remove the fasta from the end of name of file
  output = a + 'aac.csv'  #add aac.csv to the end of name of file 
  df_out = aac_wp(input, output)  #aac_wp function get input data file and calculate the aac and write in the output file 
  df_in = pd.read_csv(output)
  return df_in

feature1=aac('SPB_dataset_sequences.fasta')
feature1
```

```python
#Physico-Chemical Properties Composition
from Pfeature.pfeature import pcp_wp

def pcp(input):
  a = input.rstrip('fasta')
  output = a + 'aac.csv'
  df_out = pcp_wp(input, output)
  df_in = pd.read_csv(output)
  return df_in


feature2=pcp('SPB_dataset_sequences.fasta')
feature2
```

```python
all_feature = pd.concat([feature1, feature2], axis=1)
all_feature
```

```python
X= feature2.to_numpy()  # use only feature 2 as X
Y= y.numpy()
```

```python
# Data split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state =42)
```
